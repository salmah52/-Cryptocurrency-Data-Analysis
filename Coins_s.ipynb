```python
#This code is using the pycoingecko library to retrieve market data for Bitcoin in USD over the past 30 days from the CoinGecko API.

from pycoingecko import CoinGeckoAPI
#create cinet object
cg = CoinGeckoAPI()

#getting data in bitcoin, in usd for the past 30 days
bitcoin_data = cg.get_coin_market_chart_by_id(id='bitcoin', vs_currency = 'usd', days= 30)

```

**Here's a breakdown of the code**:

First, the pycoingecko library is imported and an instance of the CoinGeckoAPI object is created using the CoinGeckoAPI() constructor.

Next, the get_coin_market_chart_by_id() method of the cg object is called with three parameters:

**id:** This is the identifier of the cryptocurrency we want to retrieve market data for. In this case, we're using 'bitcoin' to retrieve data for Bitcoin.

**vs_currency:** This is the currency we want to get the market data in. In this case, we're using 'usd' to get the market data in US dollars.

**days:** This is the number of days of market data we want to retrieve. In this case, we're retrieving data for the past 30 days.

The get_coin_market_chart_by_id() method returns a dictionary object containing the market data for the specified cryptocurrency in the specified currency. The dictionary contains two keys: 'prices' and 'market_caps'.

The 'prices' key maps to a list of lists, where each inner list contains two elements: a Unix timestamp representing the time of the data point, and the price of the cryptocurrency at that time.

The 'market_caps' key maps to a list of lists, where each inner list contains two elements: a Unix timestamp representing the time of the data point, and the market capitalization of the cryptocurrency at that time.

You can access this data using dictionary indexing. For example, to get the prices data, you can use bitcoin_data['prices'].

```python
#create a  Pandas DataFrame called data using the pd.DataFrame constructor
# passing in the bitcoin_price_data and column names as arguments.


data = pd.DataFrame(bitcoin_price_data, columns=['TimeStamp', 'Price'])
data
```

```python
#convert to readable time series
#append to date
data['Date'] = pd.to_datetime(data['TimeStamp'], unit = 'ms')
```

```python
#in minute seconds

data['Date']
data
```

```python
#aggregate the data by date

#"min" computes the minimum value of the "Price" column for each group.
#"max" computes the maximum value of the "Price" column for each group.
#"first" computes the first value of the "Price" column for each group.
#"last" computes the last value of the "Price" column for each group.
#By using a dictionary to specify the column and the aggregation functi
candlestick_data = data.groupby(data.Date.dt.date).agg({'Price' : ["min", "max", "first", "last"]} )
...
#candlestick_data = data.groupby(data.Date).agg({'Price' : ["min", "max", "first", "last"]} ) """
```

```python
#visualise Bitcoin Candlestick Chart Over Past 30 Days
import plotly.graph_objs as go
import plotly.offline as pyo

# Rest of the code goes here

fig = go.Figure(data=[go.Candlestick(x=candlestick_data.index,
                                      open=candlestick_data['Price']['first'],
                                      high=candlestick_data['Price']['max'],
                                      low=candlestick_data['Price']['min'],
                                      close=candlestick_data['Price']['last'])
])

fig.update_layout(xaxis_rangeslider_visible=False, xaxis_title='Date',
                   yaxis_title='Price (USD $)', title='Bitcoin Candlestick Chart Over Past 30 Days')

pyo.plot(fig, filename='bitcoin_candlestick_graph.html')
```
